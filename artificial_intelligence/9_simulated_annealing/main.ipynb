```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
# допустим у нас будет функция f(x)=(x−3)^2 и как видим минимумом является 3 
def f(x):
    return x**2 + 5*np.sin(3*x) + 3*np.sin(5*x)
```

```python
# строим график фукнции

# создаем много точек
x = np.linspace(-5, 5, 400)

# считаем значения функции
y = f(x)
plt.figure(figsize=(8,5))
plt.plot(x, y)

plt.title("Функция с несколькими локальными минимумами")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.show()
```

```python
current_x = random.uniform(-5, 5)
current_y = f(current_x)

print(f"Старт  x={current_x}, y={current_y}")
```

```python
# конфигурации

T = 10 # температура
alpha = 0.99 # скорость охлаждение
ITERATION_COUNT = 1000 # количество итераций
```

```python
for i in range(ITERATION_COUNT):
    new_x = current_x + random.uniform(-0.5, 0.5)
    new_y = f(new_x)
    
    if i % 10 == 0: print(f"x={new_x}, y={new_y}")

    delta = new_y - current_y

    if delta < 0:
        current_x, current_y = new_x, new_y
    else:
        probability = math.exp(-delta/T)

        if random.random() < probability:
            current_x, current_y = new_x, new_y

    T *= alpha
```

```python
print("Найденный минимум:", current_x, current_y)
```
