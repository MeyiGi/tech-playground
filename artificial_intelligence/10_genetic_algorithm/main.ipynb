```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
# допустим у нас будет функция f(x)=(x−3)^2 и как видим минимумом является 3 
def f(x):
    return x**2 + 5*np.sin(3*x) + 3*np.sin(5*x)
```

```python
# строим график фукнции

# создаем много точек
x = np.linspace(-5, 5, 400)

# считаем значения функции
y = f(x)
plt.figure(figsize=(8,5))
plt.plot(x, y)

plt.title("Функция с несколькими локальными минимумами")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()

plt.show()
```

```python
POPULATION_COUNT = 100
GENERATION_COUNT = 5
MUTATION_RATE = 0.3
POPULATION_RANGE = 100
population = [random.uniform(-POPULATION_RANGE, POPULATION_RANGE) for i in range(POPULATION_COUNT)]
```

```python
for generation in range(1, GENERATION_COUNT + 1):
    population = sorted(population, key=f)

    print(f"Поколение {generation}: {population}")
    print(f"Лучший: {population[0]:.2f}: {f(population[0]):.2f}")
    
    parents = population[:20]

    new_population = parents.copy()

    while len(new_population) < POPULATION_COUNT:
        p1 = random.choice(parents)
        p2 = random.choice(parents)

        child = (p1 + p2) / 2

        if random.random() < MUTATION_RATE:
            child += random.uniform(-0.05 * POPULATION_RANGE, 0.05 * POPULATION_RANGE)

        new_population.append(child)

    population = new_population
```
